Pandas is a powerful and popular Python library used for data analysis and manipulation. It helps you work with large amounts of data easily, providing tools to clean, explore, and process information. With Pandas, you can load data from different sources like Excel files or CSV files and organize it into tables, making it simple to view and analyze. It offers many functions to filter, group, and sort data, making it an essential tool for data scientists, analysts, and anyone working with data. Whether you want to calculate averages, handle missing values, or prepare data for machine learning, Pandas makes data tasks much easier and faster.

```python
#1) Loads data from a CSV file into a DataFrame, making it accessible for analysis and manipulation in pandas.

import pandas as pd
df = pd.read_csv('user_behavior_dataset.csv')
```

```python
#2) 2. DataFrame.head()
# Displays the first few rows of the DataFrame, helping quickly inspect data structure and contents.

df.head()
```

```python
#3) df.info()
# Provides an overview of the DataFrame, showing column names, data types, and non-null value counts.

df.info()
```

```python
#4) df.describe()
# Generates summary statistics (mean, min, max, etc.) for numeric columns, giving insight into data distribution.

df.describe()
```

```python
#5) df.isnull()
# Checks for missing values in the DataFrame, returning a DataFrame with Boolean values (True for missing).

df.isnull()
```

```python
#6) df.fillna()
# Fills missing values in specified columns with a given value, improving data completeness for analysis.

df.fillna(0)
```

```python
#7) df.dropna()
#Removes rows or columns containing missing values, making data cleaner and more manageable for further processing.

df.dropna()
```

```python
#8) df.groupby()
# Groups data by specified columns, enabling grouped operations like aggregations or transformations on subsets of data.

grouped_df = df.groupby(['Operating System'])

print(grouped_df['Number of Apps Installed'].mean())
```

```python
#9 f.apply()
# Applies a function to DataFrame rows or columns, allowing complex transformations across data elements.

# Sample categorization function
def categorize_app_usage(usage_time):
    if usage_time < 200:
        return "Low"
    elif 200 <= usage_time <= 300:
        return "Moderate"
    else:
        return "High"

# Applying the function to the 'App Usage Time (min/day)' column
df['App Usage Category'] = df['App Usage Time (min/day)'].apply(categorize_app_usage)

# Display the updated DataFrame
df[['User ID', 'App Usage Time (min/day)', 'App Usage Category']].head()
```

```python
#10) df.merge()
# Combines two DataFrames based on common columns or indexes, similar to SQL joins for relational data.

# Additional DataFrame with device info
device_info_df = pd.DataFrame({
    'Device Model': ['Google Pixel 5', 'OnePlus 9', 'Xiaomi Mi 11'],
    'Average Battery Capacity (mAh)': [4000, 4500, 4600]
})

# Merge on 'Device Model' column
merged_df = df.merge(device_info_df, on='Device Model', how='left')

# Display the merged DataFrame
merged_df[['User ID', 'Device Model', 'Battery Drain (mAh/day)', 'Average Battery Capacity (mAh)']].head()
```

```python
#!1) df.pivot_table()
# Creates a pivot table, summarizing data by aggregating values across specified columns and indexes.

# Creating a pivot table to analyze average app usage time and battery drain
pivot_table = df.pivot_table(
    values=['App Usage Time (min/day)', 'Battery Drain (mAh/day)'],
    index='Operating System',
    columns='Gender',
    aggfunc='mean'
)

# Display the pivot table
pivot_table
```

```python
# 12) df.sort_values()
# Sorts DataFrame rows by one or more columns, helping organize data for easier analysis.

# Sorting the DataFrame by 'App Usage Time (min/day)' in descending order
sorted_df = df.sort_values(by='App Usage Time (min/day)', ascending=False)

# Display the sorted DataFrame with relevant columns
sorted_df[['User ID', 'Device Model', 'App Usage Time (min/day)', 'Battery Drain (mAh/day)']].head()
```

```python
#13) df.drop()
# Removes specified rows or columns, enabling selective deletion of unwanted data for focused analysis.

# Dropping 'Screen On Time (hours/day)' and 'User Behavior Class' columns
df_dropped = df.drop(columns=['Screen On Time (hours/day)', 'User Behavior Class'])

# Display the DataFrame with the dropped columns
df_dropped.head()
```

```python
#14) df.duplicated()
# Identifies duplicate rows in the DataFrame, helping manage and clean redundant data entries.

# Checking for duplicate rows in the DataFrame
duplicates = df.duplicated()

# Display rows that are duplicates
df[duplicates]
```

```python
# 15) df.drop_duplicates()
# Once the duplicated are identified, you can remove them using this function.

# Removing duplicate rows
df_no_duplicates = df.drop_duplicates()

# Display the DataFrame without duplicates
df_no_duplicates.head()
```

```python
# 16. df.corr()
# Calculates pairwise correlation between columns, revealing potential relationships among variables in the data.

# Selecting only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculating the correlation matrix
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
correlation_matrix
```

```python
#17) df.plot()
# Creates basic visualizations like line or bar plots, leveraging matplotlib to quickly visualize data trends.

df['Gender'].value_counts().plot(kind='pie')
```

```python
#18) df.loc[]
# Selects data by labels, allowing flexible row and column selection based on named index values.

# Using loc to select rows by User ID and specific columns
selected_data = df.loc[2:5, ['Device Model', 'App Usage Time (min/day)', 'Battery Drain (mAh/day)']]

# Display the selected data
selected_data
```

```python
# 19) df.iloc[]
# Selects data by integer location, enabling indexing and slicing rows and columns by position.

# Using iloc to select by row and column positions
selected_data_iloc = df.iloc[2:5, 0:4]

# Display the selected data
selected_data_iloc
```

```python
# 20) df.sample()
# Draws a random subset of rows from the DataFrame, useful for sampling or testing on smaller data portions.

# Selecting a random sample of 3 rows
sample_data = df.sample(n=3)

# Display the sampled data
sample_data
```

```python
# 21) df.value_counts()
# Counts unique values in a Series, often used for analyzing distributions in categorical data.

# Counting the frequency of each gender
gender_counts = df['Gender'].value_counts()

# Display the gender counts
gender_counts
```

```python
#22) df.melt()
# Converts a wide-format DataFrame to long format, ideal for reshaping data for visualization or analysis.

# Using melt to reshape the data
melted_df = df.melt(
    id_vars=['User ID', 'Device Model', 'Operating System'],
    value_vars=['App Usage Time (min/day)', 'Screen On Time (hours/day)', 'Battery Drain (mAh/day)'],
    var_name='Metric',
    value_name='Value'
)

# Display the melted DataFrame
melted_df.head()
```

```python
#23) df.explode()
# Expands list-like elements in a DataFrame column, creating separate rows for each list item.

# Sample data with a list-like column
df['Favorite Apps'] = [['YouTube', 'Instagram', 'WhatsApp']] * len(df)

## Display original DataFrame
display(df.head())

# Using explode to expand each app into its own row
exploded_df = df.explode('Favorite Apps')

# Display the exploded DataFrame
exploded_df[['User ID', 'Device Model', 'Favorite Apps']].head()
```

```python
# 24) df.nunique()
# Counts distinct values in each DataFrame column, useful for assessing the uniqueness within features.

# Counting the number of unique values
unique_counts = df['Device Model'].nunique()

# Display the unique counts for `Device Model` column
unique_counts
```

```python
# 25 df.unique()
# Shows distinct values in each DataFrame column, useful for assessing the uniqueness within features.

# unique values in specified column
unique_values = df['Device Model'].unique()

# Display the unique values for `Device Model` column
unique_values
```

```python
# 26) df.query()
# Filters rows based on a Boolean expression, providing SQL-like querying for DataFrame data.

# Define a threshold
usage_threshold = 300

# Use the variable in query
filtered_data_with_variable = df.query("`App Usage Time (min/day)` > @usage_threshold")

# Display the result
filtered_data_with_variable
```

```python
# 27. df.transpose()
# This function in pandas swaps the rows and columns of a DataFrame, effectively rotating it. This can be useful for viewing data differently or preparing it for certain analyses.

# Transpose the DataFrame
transposed_df = df.transpose()

# Display the transposed DataFrame
transposed_df
```

```python
# 28. df.rename()
# This function in pandas is used to rename columns or index labels in a DataFrame. This can be helpful for improving readability or aligning with naming conventions.

# Renaming specific columns
df_renamed = df.rename(columns={
    'App Usage Time (min/day)': 'Daily App Usage (minutes)',
    'Battery Drain (mAh/day)': 'Daily Battery Drain (mAh)'
})

# Display the DataFrame with renamed columns
df_renamed.head()
```

```python
# 29) df.resample()
# Aggregates time-series data into different intervals (e.g., daily, monthly) for more manageable analysis.

# Creating a 'Date' column with consecutive daily dates
df['Date'] = pd.date_range(start='2023-01-01', periods=700, freq='D')

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Resampling by month to calculate the average App Usage Time
monthly_usage = df['App Usage Time (min/day)'].resample('ME').mean()

# Display the resampled data
monthly_usage
```

```python
# 30) df.to_csv()
# This function in pandas allows you to save a DataFrame as a CSV file, which is helpful for exporting processed data.

# Saving the DataFrame to a CSV file
df.to_csv('processed_data.csv', index=False)
```
